## Prefix Sum
> 배열의 구간 합을 빠르게 계산하기 위해 미리 이전까지의 누적된 합을 저장해두는 알고리즘


### prefix sum 배열이란?
prefix[i] = 배열의 0번부터 i번까지의 누적합

nums    = [2, 3, 5, 1, 6]
prefix  = [2, 5, 10, 11, 17]

```python
# prefix sum 구하는 방법
nums = [2, 3, 5, 1, 6]

prefix = [0] * len(nums)
prefix[0] = nums[0]

for i in range(1, len(nums)):
    prefix[i] = prefix[i-1] + nums[i]

prefix
```

### 1732. Find the Highest Altitude
There is a biker going on a road trip. The road trip consists of n + 1 points at different altitudes. The biker starts his trip on point 0 with altitude equal 0.

You are given an integer array gain of length n where gain[i] is the net gain in altitude between points i​​​​​​ and i + 1 for all (0 <= i < n). Return the highest altitude of a point.

```python
from typing import List

class Solution:
    def largestAltitude(self, gain: List[int]) -> int:
        start, max_high = 0, 0

        for i in gain:
            start += i

            max_high = max(start, max_high)
        
        return max_high
```

### 724. Find Pivot Index
Given an array of integers nums, calculate the pivot index of this array.

The pivot index is the index where the sum of all the numbers strictly to the left of the index is equal to the sum of all the numbers strictly to the index's right.

If the index is on the left edge of the array, then the left sum is 0 because there are no elements to the left. This also applies to the right edge of the array.

Return the leftmost pivot index. If no such index exists, return -1.

```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        total = sum(nums)
        prefix = [0] * len(nums)
        prefix[0] = nums[0]

        for i in range(1, len(nums)):
            prefix[i] = nums[i] + prefix[i-1]

        
        for p in range(len(nums)):
            right = total - prefix[p]
            left = total - right - nums[p]

            if right == left:
                return p
            
        return -1



'''
[1,2,3,4,5]
sum = 15
i = 0
left = 15 - nums[i] - nums[i+1:]
right = 15 - nums[i] - left

i = 1
left = 15 - nums[i] - nums[i+1:]
right = 15 - nums[i] - left

'''
```

[위의 코드 개선 포인트]
1. prefix 배열 불필요
- 반복문 돌면서 left_sum 만 관리하면 됨
2. right, left 계산이 복잡함
- left_sum = total - left_sum - nums[i] 조건이면 됨

```python
# 개선 코드
## right_sum = total - left_sum - nums[i]

class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        total = sum(nums)
        left_sum = 0

        for i, num in enumerate(nums):
            if left_sum == total - left_sum - num:
                return i
            left_sum += num
        
        return -1
```
