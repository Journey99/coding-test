## Hash Map & Set

### hash란?
> 어떤 값을 고정된 크기의 숫자로 빠르게 매핑하는 함수
> 예를 들어 문자열 'abc' 를 숫자 129381923 같은 값으로 바꿔버리는 것

### hash map이란?
> Key-value 형태로 데이터를 저장하는 자료구조
> 파이썬에서는 dict 형태
> 언제 사용 ? 어떤 값을 빠르게 찾고 싶을 때, 빈도 계산, 두 값의 관계 저장, 맵핑 문제

### hast set이란?
> 중복 없는 값들의 집합
> 파이썬에서는 set 형태
> 언제 사용 ? 방문 여부 체크, 중복 확인, 존재 여부 검색, unique 값 관리

### 2215. Find the Difference of Two Arrays
Given two 0-indexed integer arrays nums1 and nums2, return a list answer of size 2 where:

- answer[0] is a list of all distinct integers in nums1 which are not present in nums2.
- answer[1] is a list of all distinct integers in nums2 which are not present in nums1.

Note that the integers in the lists may be returned in any order.


```python
from typing import List

class Solution:
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        nums1 = set(nums1)
        nums2 = set(nums2)

        return [list(nums1-nums2), list(nums2-nums1)]
        
```

### 1207. Unique Number of Occurrences
Given an array of integers arr, return true if the number of occurrences of each value in the array is unique or false otherwise.

```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        setarr = set(arr)
        cnt = []

        for s in setarr:
            cnt.append(arr.count(s))
        
        if len(cnt) == len(set(cnt)):
            return True
        else:
            return False
        

s = Solution()
print(s.uniqueOccurrences([1,2,2,1,1,3]))
```

```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        s = sorted(arr)

        dic = {}
        for item in s:
            dic[item] = dic.get(item, 0) + 1

        cnt = []
        for k, v in dic.items():
            if v not in cnt:
                cnt.append(v)
            else:
                return False
        
        return True
```
