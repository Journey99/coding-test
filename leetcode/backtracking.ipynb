## Backtracking
> 해를 찾기 위해 모든 경우를 탐색하지만, 유망하지 않은(답이 될 수 없는) 경우는 더 이상 진행하지 않고 바로 되돌아오는 기법
> 1. dfs로 깊이 들어가며 선택한다
> 2. 이 선택은 정답이 될 가능성이 없다 판단되면 -> 그 아래 있는 모든 경우를 탐색할 필요 없이 즉시 돌아온다
> 3. 다른 선택지로 이동하며 전체 경우를 살핀다

### 완전탐색 vs 백트래킹
- 완전탐색 : 모든 경우를 끝까지 다 본다
- 백트래킹 : 중간에 가능성 없으면 가지치기해서 속도 향상

### 백트래킹 기본 구조

```python
def backtrack(path):

    # 1. 종료 조건
    if 정답이 완성됨:
        기록
        return
    
    # 2. 가능한 선택지 순회
    for 선택 in 가능한_선택지:
        if 가지치기 조건:
            continue

        path.append(선택)
        
        # 다음단계로 진행
        backtrack(path)

        # 되돌리기 (백트래킹 핵심)
        path.pop()
```

### 백트래킹에서 가장 중요한 부분
1. 가지치기
2. 재귀 구조 설계
- 문제를 한 단계씩 내려가며 해를 구성하는 식으로 설계해야 한다
3. 상태 복원
- 선택했다가 다시 되돌리는 과정이 반드시 필요하다

### 17. Letter Combinations of a Phone Number
Given a string containing digits from 2-9 inclusive, return all possible letter combinations that the number could represent. Return the answer in any order.

A mapping of digits to letters (just like on the telephone buttons) is given below. Note that 1 does not map to any letters.



```python
from typing import List

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []
        
        buttons = {
            "2": "abc", "3": "def", "4": "ghi", "5": "jkl",
            "6": "mno", "7": "pqrs", "8": "tuv", "9": "wxyz"
        }
        
        result = []

        def backtrack(idx, path):
            if idx == len(digits):
                result.append(''.join(path))
                return
            
            for ch in buttons[digits[idx]]:
                path.append(ch)
                backtrack(idx + 1, path)
                path.pop()

        backtrack(0, [])
        return result
```

### 216. Combination Sum III
Find all valid combinations of k numbers that sum up to n such that the following conditions are true:

- Only numbers 1 through 9 are used.
- Each number is used at most once.

Return a list of all possible valid combinations. The list must not contain the same combination twice, and the combinations may be returned in any order.



```python
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:

        result = []

        def backtrack(start, path, curr_sum):
            if len(path) == k and curr_sum == n:
                result.append(path.copy())
                return

            if curr_sum > n:
                return
            
            if len(path) + (9 - start + 1) < k:
                return

            for num in range(start, 10):
                path.append(num)
                backtrack(num+1, path, curr_sum+num)
                path.pop()

               
        backtrack(1, [], 0)
        return result
```
